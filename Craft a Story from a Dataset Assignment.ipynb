### After exploring the dataset, a couple of questions come to mind: What's the average age per country? Do certain races or nationalities earn more than others? Let's investigate!

```python
country_population = income_dataset.groupby('native-country')['hours-per-week'].count().sort_values(ascending = True)
print(country_population)
```

```python
native_country_age = income_dataset.groupby('native-country')['age'].mean().sort_values()
native_country_age.plot(kind='bar', title='Median Age Per Country', ylabel='Median Age',
         xlabel='Countries', figsize=(8, 7))
```

### From this graph, we see the average worker age in Hungary is the highest. Let's see exactly how old!

```python
native_country_age.max()
```

50.4 years! Impressive! Hungary has a median age of 41.8 years, the 31st highest in the world, which may explain the particularly high age.

```python
import matplotlib.pyplot as plt

race_age = income_dataset.groupby('race')['age'].mean()
race_age.plot(kind='bar', title='Average Age Per Race', ylabel='Average Age',
         xlabel='Race', figsize=(8, 7))
```

Whites, on average, are the oldest age from this dataset. The average age of the White category looks verbatim to the United States category in the previous graph. Let's compare them.

```python
print('The average age of US workers is', native_country_age['United-States'])
print('The average age of White workers is', race_age['White'])
print("Interesting! The same age. Thus, we can conclude that the majority of US workers from this dataset is white. ")
```

```python
test2 = income_dataset.groupby('race')['hours-per-week'].mean().sort_values(ascending = True)
test2.plot(kind='bar', title='Weekly laboral hours per race', ylabel='Average hours worked per week',
         xlabel='Race', figsize=(6, 5))
print(test2)
```
